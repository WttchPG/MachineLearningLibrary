```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0.01, 1, 100)
y = -np.log2(X)
y1 = X * y

plt.plot(X, y, color='r')
plt.plot(X, y1, color='b')
```

```python
X = np.linspace(-10, 10, 201, endpoint=True)


def sigmoid(X: np.ndarray):
    return 1 / (1 + np.exp(-X))


y = sigmoid(X)

dy = y * (1 - y)

plt.plot(X, y)
plt.plot(X, dy)
```
