```python
import matplotlib.pyplot as plt
import numpy as np
from logistics.utils import sigmoid
```

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])

# 计算 s 型值
sigmoid(X_train)
```

```python
pos = y_train == 1
neg = y_train == 0

_, ax = plt.subplots(1, 1) # type: plt.Figure, plt.Subplot

ax.scatter(X_train[pos, 0], X_train[pos, 1], marker='x', s=80, c='r', label='y=1')
ax.scatter(X_train[neg, 0], X_train[neg, 1], marker='o', s=80, facecolors='none', edgecolors='#0096FF', lw=3, label='y=0')

ax.legend(loc='best')
ax.figure.canvas.toolbar_visible = False
ax.figure.canvas.header_visible = False
ax.figure.canvas.footer_visible = False
ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$', fontsize=12)
ax.set_xlabel('$x_0$', fontsize=12)
plt.show()
```
