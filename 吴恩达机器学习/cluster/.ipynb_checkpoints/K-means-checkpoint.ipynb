```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

%matplotlib notebook

#define canvas
fig, ax = plt.subplots()
line, = ax.plot([],[])

#get line 
def line_space(B):
    x = np.linspace(0,10,100)
    return x,x+B


def update(B):
    print(B)
    ax.set_xlim(0,10)
    ax.set_ylim(0,20)
    x,y = line_space(B)
    line.set_data(x,y)
    return line

ani = FuncAnimation(fig, update, frames = np.linspace(0, 20, 100), interval = 1000)
```
